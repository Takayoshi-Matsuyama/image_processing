# Split Multi-dot Image

画像認識の練習の素材として、複数点の画像を1点毎に分割する。

## 1. 分割元画像の読み込み

```python
import cv2 as cv
import matplotlib.pyplot as plt

image = cv.imread('data/multi_dot_image1.jpg')  # 読み込み
plt.imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB))  # 表示。BGR -> RGBへ変換
plt.show()
```

## 2. 画像の切り出し

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

def crop_image(image: np.ndarray, left: int, top: int, width: int = 400, height: int = 400) -> np.ndarray:
    """画像を切り取る関数

    Args:
        image (np.ndarray): 切り取る対象の画像
        left (int): 切り取る領域の左端のx座標
        top (int): 切り取る領域の上端のy座標
        width (int): 切り取る領域の幅
        height (int): 切り取る領域の高さ

    Returns:
        np.ndarray: 切り取られた画像
    """

    cropped_image = image[top:top+height, left:left+width]  # 終点は含まない
    plt.imshow(cv.cvtColor(cropped_image, cv.COLOR_BGR2RGB))
    return cropped_image
```

```python
cropped_image_01 = crop_image(image, left=460, top=690)
```

## 3. 切り出した画像を保存

```python
import cv2 as cv

cropped_image_list = [
    cropped_image_01,
    cropped_image_02,
    cropped_image_03,
    cropped_image_04,
    cropped_image_05,
    cropped_image_06,
    cropped_image_07,
    cropped_image_08,
    cropped_image_09,
    cropped_image_10,
    cropped_image_11,
    cropped_image_12
]

for i, cropped_image in enumerate(cropped_image_list, start=1):
    # single_dot_image01.jpg, single_dot_image02.jpg, ..., single_dot_image12.jpg
    cv.imwrite(f'./data/single_dot_image{i:02d}.jpg', cropped_image)
```
